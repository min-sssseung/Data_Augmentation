```python
from PIL import Image, ImageEnhance, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import pdb
from autoaugment import ImageNetPolicy, CIFAR10Policy, SVHNPolicy, SubPolicy
```

```python
img = Image.open("Fire_image.png")
plt.imshow(img)
```

```python
policy = ImageNetPolicy()

def show_imgs(img):
    f, axarr = plt.subplots(3, 3, figsize=(6, 6))
    for idx, ax in enumerate(f.axes):
        ax.imshow(img[idx])
        ax.axis("off")
    plt.show()

imgs = []
for _ in range(9):
    imgs.append(policy(img))

show_imgs(imgs)
```

```python
policy = CIFAR10Policy()

imgs = []
for _ in range(9):
    imgs.append(policy(img))

show_imgs(imgs)
```

```python
policy = SVHNPolicy()

imgs = []
for _ in range(9):
    imgs.append(policy(img))

show_imgs(imgs)
```
